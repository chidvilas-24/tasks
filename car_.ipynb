```python
def cyl(n):
    if n == 'two':
        return 2
    elif n== 'three':
        return 3
    elif n == 'four':
        return 4
    elif n == 'five':
        return 5
    elif n == 'six':
        return 6
    elif n == 'eight':
        return 8
    else:
        return 12
```
